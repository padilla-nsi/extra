# Timer, part 2


## Menu de configuration

### Nouveau panneau

- créer un nouveau `Panel` enfant de `Control`
- renommer le `ConfigMenu`
- dans le menu `Disposition sur l'écran` choisir de le positionner `En bas à gauche` puis déplacer le avec la souris pour le mettre sous le panneau `Main`
- Pour le colorier :
  - dans `Theme Overrides > Styles Panel` créer un `Nouveau StyleBoxFlat`
  - cliquer sur `bg color` puis choisir la couleur foncée avec la pipette
  - positionner le pointeur de couleur sur une teinte plus claire, par exemple `631b6e`

### Contenu de la barre d'en-tête : icône 

- créer un node `ColorRect` enfant du node `HBoxContainer`
  - modifier la taille `Size` en 150 de large (`x`) et 150 de haut (`y`)
  - modifier la taille minimale `Min Size` en 150 par 150 aussi
  - modifier la couleur en choisissant la couleur de fond du panneau `ConfigMenu`
  - renommer ce node en `icoBack` (mais on le renommera autrement un peu plus tard...)
- télécharger l'icône qui va bien sur google fonts
  - chercher `back`
  - télécharger au format `png` en ayant au préalable configuré la densité, la taille et la couleur
  - dézippé dans le dossier de travail
- revenir sur Godot
- sélectionner le node `IcoBack`
- faire glisser l'image `icoBack.png` téléchargée vers le centre du node `IcoBack`
  - configurer la `Position` sur le centre de coordonnées (75 ; 75)

### Contenu de la barre d'en-tête : le texte

### Script pour détecter le passage de la souris sur l'icône

- ajouter un nouveau `Script` au node `IcoBack`
  - valider la création du fichier `IcoBack.gd`
- connecter le signal `mouse_entered()` au nœud `IcoBack` (attention, la capture d'écran montre le mauvais signal)
  - sélectionner le script de `IcoBack` et cliquer sur `Connecter`
- En utilisant l'outil pipette, récupére le code **hexadécimal**  de la couleur rose (code qui doit être `d94bc9`)
- Mettre à jour la fonction `_on_IcoBack_mouse_entered()`
- Le programme détecte correctement le survol de la souris mais ne revient pas à la couleur d'origine

- connecter le signal `mouse_exited()` au nœud `IcoBack`
  - sélectionner le script de `IcoBack` et cliquer sur `Connecter`
- Mettre à jour la fonction `_on_IcoBack_mouse_exited()`
- Le programme détecte correctement le survol de la souris et revient à la couleur d'origine

## Bouton d'accueil : Configuration

### Modification du bouton de configuration

- supprimer le node `ParameterButton`
- ajouter le node `ColorRect` enfant de `Control > Main`
  - modifier sa taille `Size` en 150 par 150
  - choisir la `Disposition sur l'écran` de type `En bas à gauche`
  - renommer ce nœud en `ConfigButton`
  - (et pour homogénéiser les notations, renommer `IcoBack` en `BackButton`)
  - changer la couleur du node `ConfigButton`
    - utiliser la pipette pour choisir la couleur de fond du panneau principal `Main` (normalement la couleur est `29082e`)
  - télécharger l'icône de menu
    - chercher `menu` sur google fonts
    - télécharger au format PNG (en ayant préalablement configuré les paramètres de l'image)
    - dézipper dans le dossier de travail et renommer le fichier PNG en `icoConfig.png`
  - revenir sur Godot
  - sélectionner le node `ConfigButton`
  - faire glisser l'image `icoConfig.png` vers le centre de ce nœud
- sélectionner `IcoConfig`
  - modifier sa `Position` pour le placer au centre précisément en choisissant (75 ; 75) comme coordonnées

### Survol de la souris

- sélectionner `ConfigButton`
  - créer un nouveau script (attention, pas de capture d'écran)
  - dans les signaux :
    - connecter le signal `mouse_entered()` au script `ConfigButton.gd` 
    - connecter le signal `mouse_exited()` au script `ConfigButton.gd`
- dans le script `ConfigButton.gd` :
  - modifier les fonctions `_on_ConfigButton_mouse_entered()` et `_on_ConfigButton_mouse_exited()`

### Détection du clic de la souris

- sélectionner le node `Control`
  - le renommer en `App`
  - créer un nouveau `Script`
  - et valider la création du fichier `main.gd`
- sélectionner le node `ConfigButton`
  - choisir le signal `gui_input(event)` et
  - attention à bien sélectionner `App` pour connecter le bouton au script principal
- Mettre à jour le code du fichier `main.gd`

Désormais l'application détecte les différents clics de souris et simule l'affichage du panneau configuration et sa disparition.
