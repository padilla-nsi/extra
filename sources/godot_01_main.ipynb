# Timer

## Fenêtre MAIN


### Paramètre du projet

- Nouveau projet
- Ajouter un `Control`
- Changer les paramètres du projets
    - changer la taille : 720 × 1280
    - diminué le FPS : 30 images/s


![](./2021-12-02/img_1.png)

![](./2021-12-02/img_2.png)


### Panneau conteneur et titre de la fenêtre

- Ajouter un `Panel` (panneau) et le renommer `Main`
- Ajouter dans `Main` un `Label`
  - écrire le texte `écran MAIN`
  - ajouter la police d'écrite `Roboto-Regular.ttf` (*font* en anglais) dans le dossier du projet
  - Ajouter un nouveau thème
  - Dans le thème, ajouter dans `Default Font` une `DynamicFont`
  - Faire glisser la police dans `Font Data`

![](./2021-12-02/img_3.png)

![](./2021-12-02/img_4.png)


  - changer la taille de la police
  - Enregistrer le thème

![](./2021-12-02/img_5.png)

![](./2021-12-02/img_6.png)


### Texte du timer

- Ajouter dans `Main` un deuxième `Label`
  - renommer ce Label en `textTimer`
  - changer le texte en `9999`
  - Appliquer le thème sauvegardé dans la partie précédente
  - Disposer ce label au centre du panneau

![](./2021-12-02/img_7.png)


### Google fonts : polices et icones

- Visiter le site 'google fonts' : https://fonts.google.com
- Dans la barre de recherche, écrire *para*
  - sélectionner l'icône recherchée
  - changer sa taille : size => 48 
  - sa couleur : color => white
  - cliquer sur `SVG` pour télécharger
  - enregistrer le fichier dans le dossier du projet
  - renommer le fichier en un nom plus explicite : `icoMenu.svg`

![](./2021-12-02/img_8.png)

![](./2021-12-02/img_9.png)



### Ajoute l'icône dans Main

- Dans le panel `Main` ajouter un Node `TextureButton`
  - faire glisser l'icône téléchargée dans le menu Textures > Normal du `TextureButton`
  - étirer le `TexturButton` pour l'afficher deux fois plus grand en modifiant la propriété `Scale` de `Rect` de 1 à la nouvelle valeur 2 (pour `x` et `y`)


![](./2021-12-02/img_10.png)

![](./2021-12-02/img_11.png)

![](./2021-12-02/img_12.png)

![](./2021-12-02/img_13.png)


  - Positionner l'icône au bon endroit
    - Modifier le point d'ancrage du bouton
      - Choisir dans `Disposition sur l'écran` le choix intitulé `En bas à gauche`
      - Ajuster la position du bouton

![](./2021-12-02/img_18.png)



### Choix des couleurs pour le panneau

- Sélectionner le panel `Main`
  - Dans `Theme override` > `Style`, ajouter un  *Nouveau* `StyleBoxFlat`
  - Cliquer sur `Bg color` (*background color*, couleur de fond) pour modifier la couleur du panneau.
    - choisir : `29082e` (violet)


![](./2021-12-02/img_14.png)

![](./2021-12-02/img_15.png)


- Dans les propriétés du projet
  - Rechercher le terme `color`
  - Sélectionner `Rendering` > `Environment`
    - Modifier la `Default Clear Color`
    - Choisir : `d94bc9`
